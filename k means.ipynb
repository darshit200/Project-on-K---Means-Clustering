```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Define the dataset 
data_points = np.array([
    [2, 10], [2, 5], [8, 4], [5, 8], [7, 5], 
    [6, 4], [1, 2], [4, 9]  
])

# Define initial centroids 
initial_centroids = np.array([[2, 10], [5, 8], [1, 2]])

# Perform K-Means clustering with predefined centroids
kmeans = KMeans(n_clusters=3, init=initial_centroids, max_iter=100, n_init='auto', random_state=42)
kmeans.fit(data_points)

# Get final cluster assignments and centroids
labels = kmeans.labels_
final_centroids = kmeans.cluster_centers_

# Print results
print("\nFinal Cluster Assignments:")
for i, point in enumerate(data_points):
    print(f"Point {point} -> Cluster {labels[i] + 1}")

print("\nFinal Centroids:\n", final_centroids)

# Define cluster colors
colors = ['red', 'green', 'blue']  

# Visualization
plt.figure(figsize=(6, 6))

for i in range(3):
    cluster_points = data_points[labels == i]  # Ensure proper selection of points
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], c=colors[i], label=f'Cluster {i + 1}')

# Plot final centroids
plt.scatter(final_centroids[:, 0], final_centroids[:, 1], c='black', marker='X', s=200, label='Final Centroids')

plt.xlabel("X Coordinate")
plt.ylabel("Y Coordinate")
plt.title("K-Means Clustering (Updated Colors)")
plt.legend()

plt.savefig("kmeans_clusters.png", dpi=300)

plt.show()

print("Clustering image saved as 'kmeans_clusters.png'.")
```
